```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the csv file :-
df = pd.read_csv('/content/Diwali Sales Data.csv', encoding='unicode_escape') #if there is errorin csv file opening use unicode_escape encoding
df.shape
```

```python
df.head(10)
```

```python
df.info() #provides full information about the data
```

```python
#to drop the unnecessary columns from the data we use the dop function :-
df.drop(['Status','unnamed1'], axis =1 , inplace = True)
```

```python
df.info()
```

```python
#Checking if there are any null values :-
pd.isnull(df).sum()
```

```python
#to drop the null values we use the dropna function :-
df.dropna(inplace = True)
```

```python
df.shape
```

```python
#changing the datatype of column :-
df['Amount'] = df['Amount'].astype('int')
```

```python
#for changing the column name for the data use rename function :-
df.rename(columns={'Marital_Status' : 'Shaadi'})
```

**Exploratory** ***data*** **analysis**

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender' , data = df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending = False)
```

```python
sales_gen =df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending = False)
sns.barplot(x='Gender',y='Amount',data = sales_gen)
```

This shows that the numbers of buyers from female category were more than the male category

**Age**

```python
df.columns
```

```python
ax = sns.countplot(x='Age Group',data = df , hue = 'Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'] , as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = False)
sns.barplot(x='Age Group', y = 'Amount' , data = sales_age )
```

From the above graph it can be concluded that the age group between 26-35 are the major buyers .


**State**

```python
df.columns
```

```python
ax = sns.countplot(x ='State' , data = df , hue = 'Gender')
for bars in ax.containers:
  ax.bar_label(bars)
sns.set(rc = {'figure.figsize':(30 ,5)})
```

```python
sales_state = df.groupby(['State'] , as_index = False)["Orders"].sum().sort_values(by = 'Orders' , ascending = False).head(10)
sns.set(rc ={'figure.figsize':(20,5)})
sns.barplot(x = 'State' , y ='Orders' , data = sales_state )
```

similarly as we have done for orders we can also done for the total amount of money that is for the amount

From this we can conclude that from U.P we have got highest number of orders and from Gujarat we got less orders

**Mariatal** **Status**

```python
ax = sns.countplot(x = 'Marital_Status' , data = df)
for bars in ax.containers:
  ax.bar_label(bars)
sns.set(rc = {'figure.figsize' :(7,5)})
```

```python
sales_Marital_Status = df.groupby(['Marital_Status' ,'Gender'] , as_index = False)["Amount"].sum().sort_values(by = 'Amount' , ascending = False).head(10)
sns.set(rc ={'figure.figsize':(7,5)})
sns.barplot(x = 'Marital_Status' , y ='Amount' , data = sales_Marital_Status , hue ='Gender')
```

from this we can conclude that married females were the largest buyer
